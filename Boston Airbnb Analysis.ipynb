## During which time of year is it the most expensive?

Plot the average price per day for each calendar year

### Clean data

```python
#Remove the dollar sign from the price column
conv_str = lambda x: str(x)[1:].replace(",","") if "$" in str(x) else x

calendar2017["price"] = calendar2017["price"].apply(conv_str)
calendar2017["price"] = calendar2017["price"].apply(lambda x: float(x))
calendar2017.head()
```

```python
#Get summary statistics for the average price per listing
grouped_price = calendar2017.groupby("listing_id")["price"].mean()
grouped_price.describe()
```

```python
#Function to apply the data cleaning to any input data file
def clean(df, data_column, string_remove):
    """
    data_column - column of the data to be cleaned, string_remove - remove specified string from data
    df - Cleaned dataframe
    """
    conv_str = lambda x: str(x)[1:].replace(",","") if string_remove in str(x) else x
    df[data_column] = df[data_column].apply(conv_str)
    df[data_column] = df[data_column].apply(lambda x: float(x))
    
    return df
```

```python
#Function to produce plots
def format_plot(df):
    
    #average price throughout time
    grouped_date = df.groupby('date')['price'].mean()
    #Remove year from date column
    index_date = [x[5:] for x in grouped_date.index]
    
    return index_date, grouped_date
```

```python
# clean all calendar price columns
X2017, y2017 = format_plot(clean(calendar2017, 'price', '$'))
X2018, y2018 = format_plot(clean(calendar2018, 'price', '$'))
X2019, y2019 = format_plot(clean(calendar2019, 'price', '$'))
```

```python
# Remove year from date
X2017 = [datetime.strptime(x, '%m-%d') for x in X2017]
X2018 = [datetime.strptime(x, '%m-%d') for x in X2018]
X2019 = [datetime.strptime(x, '%m-%d') for x in X2019]
```

```python
#Create input list for plot function
x_list = [X2017, X2018, X2019]
y_list = [y2017,y2018,y2019]
year_list = ['09/2016 until 09/2017', '09/2017 until 09/2018', '09/2018 until 09/2019']
colour_list = ["b","r","g"]
```

### Analyse and Visualise Data

```python
#Plot a distribution of the price of listings in 2017
plt.figure(figsize=(15,6))
plt.ylabel("Average Price")
plt.hist(grouped_price, bins=100);
```

```python
#Plot the average price of listings across time for the year 2017
grouped_date = calendar2017.groupby("date")["price"].mean()
index_date = [np.datetime64(x) for x in grouped_date.index]

fig, ax = plt.subplots(figsize=(15,6))

plt.scatter(index_date, grouped_date.values)
plt.xlabel("Date")
plt.ylabel("Average Price")
plt.title("Average price on date throughout 2017")
plt.xticks(rotation= 30);
```

```python
#Function to plot the average price of listings for all years

def plot_cal(x_list, y_list, year_list, colour_list):
    """
    INPUT: x_list = index_date, y_list = grouped_date
    OUTPUT:
    """
        
    fig,ax = plt.subplots(figsize=(15,6))
    
    for i in range(len(x_list)):
        ax.scatter(x_list[i], y_list[i].values, c=colour_list[i], label=year_list[i])
        
    plt.xlabel("Date")
    plt.ylabel("Average Price")
    ax.set_ylim([100,500])
    monthyearFmt = mdates.DateFormatter("%B")
    ax.xaxis.set_major_formatter(monthyearFmt)
    plt.legend(loc="upper left")
    plt.title("Average prices throughout the year");
```

```python
plot_cal(x_list, y_list, year_list, colour_list)
```
